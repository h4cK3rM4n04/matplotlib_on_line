# Exercice 1:

Tracer un graphique en courbes des fonctions $f(x) = x$ et $g(x) = x^2$ pour x allant de -10 Ã  10.

```python
#coding:utf-8 by h4cK3rM4n04
import matplotlib.pyplot as pyp
import numpy as num
def f(x):
    return x
def g(x):
    return x ** 2
fig , ax = pyp.subplots()
x = num.arange(-10,10)
y = f(x)
pyp.plot(x, y, label="f(x)")
y = g(x)
pyp.plot(x, y, label="g(x)")
pyp.legend()
pyp.xlabel("Axe des (x)")
pyp.ylabel("Axe des (y)")
ax.set_facecolor("lightblue")
ax.grid()
fig.set_facecolor("darkgrey")
```

# Exercice 4:

CrÃ©er un diagramme circulaire pour reprÃ©senter les parts de marchÃ© de diffÃ©rentes entreprises dans un secteur donnÃ©.

```python
#coding:utf-8 by h4cK3rM4n04
import matplotlib.pyplot as plt
import numpy as np
figure, axe = plt.subplots()
buisiness_market = {"apple": 50, "meta": 92, "android": 35, "openai": 85, "h4cK3rM4n04": 100}
x = [x for x in buisiness_market.values()]
y = [x for x in buisiness_market.keys()]
array = np.array(y)
plt.pie(x, labels = array, autopct = lambda x:f"{x:.1f}%",
       colors = ["yellow", "blue", "green", "lightblue", "red"])#lambda x:f"{x:.1f}%" cette ligne de code peut Ãªtre remplacÃ© par %1.1f%%
axe.legend(title = "Diagramme")
plt.show()
```
